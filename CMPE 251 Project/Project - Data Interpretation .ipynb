```python
import pandas as pd
```

```python
class color:
   PURPLE = '\033[95m'
   CYAN = '\033[96m'
   DARKCYAN = '\033[36m'
   BLUE = '\033[94m'
   GREEN = '\033[92m'
   YELLOW = '\033[93m'
   RED = '\033[91m'
   BOLD = '\033[1m'
   UNDERLINE = '\033[4m'
   END = '\033[0m'
```

```python
#df_general = pd.read_csv('../CMPE 251 Project/CSV/General.csv')
df_avg_general = pd.read_csv('../CMPE 251 Project/CSV/General Avg.csv')
#df_general = df_general.drop(columns = 'Min-Max Values - 1 track')
df_avg_general = df_avg_general.drop(columns='Min-Max Values')
```

```python
class interpret(object):
    def __init__(self, *args, **kwargs):
        super().__init__(*args, **kwargs)
        
    def get_partial_data(self, df_for_row, row_name):
        partial =  df_for_row.loc[f'{row_name}',:]
        partial = pd.DataFrame(partial)
        partial.index= ['2018', '2019', '2020']
        return partial
    
    def numbers_for_name(self, name):
        output = ''
        if 'pop' in name.lower():
            output = 'popular'
        elif 'danc' in name.lower():
            output = 'danceable'
        elif 'ener' in name.lower():
            output = 'energetic'
        elif 'loud' in name.lower():
            output = 'closer to 0 dB'
        elif 'speech' in name.lower():
            output = 'speech included'
        elif 'temp' in name.lower():
            output = 'tempo'
        elif 'vale' in name.lower():
            output = 'valence'
        return output
    
    def inc_dec(self, old, new):
        if old > new:
            return 'decreased'
        else:
            return 'increased'
        return ''
    
    def adj_check(self, name, check_list):
        out = False
        counter=0
        for counter in range(0,len(check_list)):
            if check_list[counter] in name.lower():
                 return True
            else:
                counter +=1
        return False
    
    def adj(self, name, old, new):
        name = name.lower()
        more_less_list = ['pop', 'dan', 'ene' ,'loud', 'speec']
        high_low_list = ['temp', 'vale']
        
        output = self.adj_check(name, more_less_list)
        out_return = ''
       
        if output:
            if new > old:
                out_return = 'more'
            else: 
                out_return = 'less'
        else:
            output = self.adj_check(name,high_low_list)
            if output:
                if new > old:
                    out_return = 'high'
                else:
                    out_return = 'low'
        return out_return
        
        
    def compare(self, comp_data):
        status = ""
        data18 = comp_data.loc['2018',:]
        data19 = comp_data.loc['2019',:]
        data_before20 = (float(data18+data19))/2
        data20 = float(comp_data.loc['2020',:])
        name = comp_data.columns
        name = name[0]
        
        word = self.numbers_for_name(name)
        inc_or_dec = self.inc_dec(data_before20, data20)
        adj = self.adj(name,data_before20,data20)
        
        first_sentence = f'{color.UNDERLINE}{name.upper()}{color.END}'
        middle_sentence = f'{color.UNDERLINE}{color.BOLD}{color.BLUE}{inc_or_dec.upper()}{color.END}'
        last_sentence = f'{color.UNDERLINE}{color.BOLD}{color.RED}{adj.upper()} {word.upper()}{color.END}'
        
        status = f"""\rWhen the values are compared in terms of {first_sentence};
            \r\tValue of {name} is {middle_sentence} which means:
                \r\t\tUser started to listen {last_sentence} songs in 2020.
                    """;
        return status
        
```

```python
itp = interpret()
```

```python
prt_pop=itp.get_partial_data(df_avg_general , 'Avg Popularity')
prt_dance=itp.get_partial_data(df_avg_general , 'Avg Danceability')
prt_energy=itp.get_partial_data(df_avg_general , 'Avg Energy')
prt_loud=itp.get_partial_data(df_avg_general , 'Avg Loudness')
prt_speech=itp.get_partial_data(df_avg_general , 'Avg Speechiness')
prt_tempo=itp.get_partial_data(df_avg_general , 'Avg Tempo')
prt_valence=itp.get_partial_data(df_avg_general , 'Avg Valence')

comp_pop = itp.compare(prt_pop)
comp_dan = itp.compare(prt_dance)
comp_ene = itp.compare(prt_energy)
comp_lou = itp.compare(prt_loud)
comp_spe = itp.compare(prt_speech)
comp_tem = itp.compare(prt_tempo)
comp_val = itp.compare(prt_valence)

all_in_list = [comp_pop, comp_dan, comp_ene, comp_lou, comp_spe, comp_tem, comp_val]
```
